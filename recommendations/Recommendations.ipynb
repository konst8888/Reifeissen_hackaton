```python
cat_cols = [
    'categorycode',
    'gender',
    'married_',
    'residenttype'
]
real_cols = [
    'amount',
    'age'
]
```

```python
df_train = df_trans[df_trans.category == 'Clothes'][['cnum'] + cat_cols + real_cols]
df_train.isna().sum(axis=0)
```

```python
store_name = 'Louis Vuitton'

df_train = df_trans[df_trans.category == 'Clothes'][['cnum', 'store_name'] + cat_cols + real_cols]

for col in tqdm(cat_cols):
    dummies = pd.get_dummies(df_train[col])
    dummies.columns = [col + str(c) for c in dummies.columns]
    df_train = pd.concat([
        df_train,
        dummies
    ], axis=1)
    del df_train[col]
    
df_store = df_train[df_train.store_name == store_name]
del df_store['store_name']
del df_train['store_name']
```

```python
for col in df_train.columns:
    if df_train[col].dtype != float and col != 'cnum':
        df_train[col] = df_train[col].astype(float)
```

```python
train_cols = df_train.columns.to_list()
train_cols.remove('cnum')
#df_train['freq'] = df_train['cnum']
aggdict = {}
aggdict.update({col: 'mean' for col in train_cols})
#aggdict.update({'freq': 'count'})

pivot_cnum = df_train.pivot_table(
    train_cols, 
    ['cnum'], 
    aggfunc=aggdict
)
pivot_cnum.reset_index(inplace=True)
pivot_cnum = pivot_cnum[df_store.columns]
pivot_cnum.shape
```

```python
from scipy.spatial.distance import cosine

mean_vec = df_store.mean(axis=0).to_list()
#cols = list(set(df_store.columns) - set(['cnum']))
#weights = [1 / pivot_cnum[col].unique().shape[0] for col in cols]
#weights = np.array(weights) / sum(weights)
pivot_cnum['score'] = np.nan

for idx, row in tqdm(pivot_cnum.iterrows(), total=len(pivot_cnum)):
    #row = row[cols]
    pivot_cnum.at[idx, 'score'] = cosine(mean_vec, row.to_list()[1:-1], 
                                         #w=weights
    )
```
